```python
import json

def generate_cluster_text(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)

    # Sort cluster IDs numerically
    sorted_clusters = sorted(data.items(), key=lambda x: int(x[0]))

    for cluster_id, cluster_data in sorted_clusters:
        cluster_def = cluster_data["definition"]
        subclusters = cluster_data.get("subclusters", {})

        print(f"Cluster ID: {cluster_id} - {cluster_def}")

        # If more than one subcluster, print them
        if len(subclusters) > 1:
            # Sort subclusters numerically (1.1, 1.2, ...)
            sorted_subs = sorted(
                subclusters.items(),
                key=lambda x: float(x[0])
            )

            for sub_id, sub_data in sorted_subs:
                sub_def = sub_data["definition"]
                print(f"    Cluster ID: {sub_id} - {sub_def}")

        print()  # blank line between clusters


if __name__ == "__main__":
    generate_cluster_text("EB03.json")
```
